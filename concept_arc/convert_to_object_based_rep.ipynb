```python
import threading
from threading import Thread
from time import sleep
import sys
import os
import numpy as np
from sklearn.preprocessing import LabelEncoder
import json
```

```python
sys.path.append("./ARGA_AAAI23")
from ARGA_AAAI23.task import Task

root_path = "./corpus"
X = []
labels = []
threads=[]
for concept in os.listdir(root_path):
    for task in os.listdir(os.path.join(root_path, concept)):
        task_path = os.path.join(root_path, concept, task)
        X.append(None)
        labels.append(concept)
        threads.append(None)

        task = Task(task_path)

        i = len(threads)-1
        threads[i] = Thread(target=task.get_obj_representation, args=(True, 600, True, True, i, X))
        threads[i].start()

for thread in threads:
    thread.join()
    
```

```python
root_path = "./corpus"
labels = []
for concept in os.listdir(root_path):
    for task in os.listdir(os.path.join(root_path, concept)):
        task_path = os.path.join(root_path, concept, task)
        labels.append(concept)
```

```python
le = LabelEncoder()
labels = le.fit_transform(labels)
root_path = "./object_based_corpus"
```

```python
count = []
for label in np.unique(labels):
    label_name = le.inverse_transform([label])[0]
    indexes = np.where(labels == label)[0]
    os.makedirs(os.path.join(root_path, label_name))
    count.append(0)
    for idx in indexes:
        task = {}
        train = []
        if X[idx] == None: continue
        for i in range(len(X[idx][0])):
            example = {}
            example["input"] = X[idx][0][i]
            example["output"] = X[idx][1][i]
            train.append(example)

        test = []
        for i in range(len(X[idx][2])):
            test_example = {}
            test_example["input"] = X[idx][2][i]
            test_example["output"] = X[idx][3][i]
            test.append(test_example)

        task["train"] = train
        task["test"] = test

        with open(os.path.join(root_path, label_name, f"{idx}.json"), 'w') as f:
            json.dump(task, f)
        
        count[-1]+=1
        
count
```
